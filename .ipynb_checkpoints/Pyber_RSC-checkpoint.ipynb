```python
# Define all Dependencies

# Include this line for interactive plots in Matplotlib
%matplotlib inline

# All other Dependencies
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read/Load all files

# File to Load
ride_data_csvpath = os.path.join('C:', '/', 'Users', 'greg_', 'Pyber-Ride-Sharing-Co-Analysis', 'ride_data.csv')
city_data_csvpath = os.path.join('C:', '/', 'Users', 'greg_', 'Pyber-Ride-Sharing-Co-Analysis', 'city_data.csv')

# Read the City and Ride Data
ride_data_df = pd.read_csv(ride_data_csvpath)
city_data_df = pd.read_csv(city_data_csvpath)

# Combine the data into a single dataset
ride_and_city_data_df = pd.merge(ride_data_df, city_data_df, on="city", how="outer")

# Display the data table for preview
ride_and_city_data_df.head(10)

# Create, Cleanup & Prepare a df for processing for charts
city_type_group = ride_and_city_data_df.groupby(["city"])
city_type_df = pd.DataFrame(city_type_group["city"].first())
city_type_df["date"] = city_type_group["date"].first()
city_type_df["avg_fare"] = city_type_group["fare"].mean()
city_type_df["avg_fare"] = city_type_df["avg_fare"].map("{:.0f}".format)
city_type_df["ride_id"] = city_type_group["ride_id"].count()
city_type_df = city_type_df.rename(columns={"ride_id":"ride_id_count"})
city_type_df["driver_count"] = city_type_group["driver_count"].first()
city_type_df["type"] = city_type_group["type"].first()
city_type_df = city_type_df.drop(["city"], axis=1) 
city_type_df.head(10)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_df = city_type_df.loc[city_type_df["type"] == "Urban"]
suburban_df = city_type_df.loc[city_type_df["type"] == "Suburban"]
rural_df = city_type_df.loc[city_type_df["type"] == "Rural"]

# Print some preliminary data from the above
print("Some Preliminary Counts:")
print("-----------------------")
print("There are a Total of " + str(city_type_df["type"].count()) + " cities in our dataset")
print(str(urban_df["type"].count()) + " are Urban cities")
print(str(suburban_df["type"].count()) + " are Suburban cities")
print(str(rural_df["type"].count()) + " are Rural cities")

# x_axis = Total Number of Rides (Per City)
x_urban = urban_df["ride_id_count"]
x_suburban = suburban_df["ride_id_count"]
x_rural = rural_df["ride_id_count"]

# y_axis = Average Fare ($)
y_urban = urban_df["avg_fare"]
y_suburban = suburban_df["avg_fare"]
y_rural = rural_df["avg_fare"]

# Get driver count per city for bubble size (s=) on plot
# I did the "times 10" because all of the circles were too small & the plot did not look pretty
urban_dc = (urban_df["driver_count"] * 10)
suburban_dc = (suburban_df["driver_count"] * 10)
rural_dc = (rural_df["driver_count"] * 10)

# Print some more preliminary data from the above
urban_dc_fmt = format(urban_df["driver_count"].sum(), ",d")
urban_dc_min_fmt = format(urban_df["driver_count"].min(), ",d")
urban_dc_max_fmt = format(urban_df["driver_count"].max(), ",d")
urban_dc_avg_fmt = format(urban_df["driver_count"].mean(), ".1f")
suburban_dc_fmt = format(suburban_df["driver_count"].sum(), ",d")
suburban_dc_min_fmt = format(suburban_df["driver_count"].min(), ",d")
suburban_dc_max_fmt = format(suburban_df["driver_count"].max(), ",d")
suburban_dc_avg_fmt = format(suburban_df["driver_count"].mean(), ".1f")
rural_dc_fmt = format(rural_df["driver_count"].sum(), ",d")
rural_dc_min_fmt = format(rural_df["driver_count"].min(), ",d")
rural_dc_max_fmt = format(rural_df["driver_count"].max(), ",d")
rural_dc_avg_fmt = format(rural_df["driver_count"].mean(), ".1f")
print()
print("There are " + str(urban_dc_fmt) + " Urban drivers. Their numbers range from " + str(urban_dc_min_fmt) +
      " to " + str(urban_dc_max_fmt) + " per city with an average of " + str(urban_dc_avg_fmt) + ".")
print("There are " + str(suburban_dc_fmt) + " Suburban drivers. Their numbers range from " + str(suburban_dc_min_fmt) +
      " to " + str(suburban_dc_max_fmt) + " per city with an average of " + str(suburban_dc_avg_fmt) + ".")
print("There are " + str(rural_dc_fmt) + " Rural drivers. Their numbers range from " + str(rural_dc_min_fmt) +
      " to " + str(rural_dc_max_fmt) + " per city with an average of " + str(rural_dc_avg_fmt) + ".")
print()

# Incorporate the other graph properties
# x & y axis Limits
plt.xlim(0, city_type_df["ride_id_count"].max() + 2) 
city_type_df["avg_fare"] = pd.to_numeric(city_type_df["avg_fare"])
plt.ylim(city_type_df["avg_fare"].min() - 2, city_type_df["avg_fare"].max() + 2)
# Grid
plt.grid()
# Title and x & y axis labels
plt.title("Pyber Ride Sharing Data (2018)", pad=15, fontsize=20) # pad moves the title away from the top axis so it looks better
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Build the scatter plots for each city types
urban_handle = plt.scatter(x=x_urban, y=y_urban, marker="o", color="lightcoral", edgecolors="black", s=urban_dc, alpha=0.8, label="Urban")
suburban_handle = plt.scatter(x=x_suburban, y=y_suburban, marker="o", color="lightskyblue", edgecolors="black", s=suburban_dc, alpha=0.8, label="Suburban")
rural_handle = plt.scatter(x=x_rural, y=y_rural, marker="o", color="gold", edgecolors="black", s=rural_dc, alpha=0.8, label="Rural")
# Set all of the legend markers to be the same size
lgnd = plt.legend(handles=[urban_handle, suburban_handle, rural_handle], title="City Types", loc="best")
for handle in lgnd.legendHandles:
    handle.set_sizes([25.0])

# Incorporate a text label regarding circle size
plt.text(42, 40, "Note: Circle size coorelates with driver count per city")

# Save Figure
plt.savefig("ridesharingdata.png")

# Plot it
plt.show()
print()
```

## Total Fares by City Type

```python
# Calculate Type Percents
# Create, Cleanup & Prepare a df for processing for charts
tfbct_group = ride_and_city_data_df.groupby(["type"])
tfbct_df = pd.DataFrame(tfbct_group["fare"].sum())
tfbct_df["type"] = tfbct_group["type"].first()
tfbct_df["pct"] = (tfbct_df["fare"] / (tfbct_df["fare"].sum()) * 100)
tfbct_df.head()

# Create the 2 lists for the pie chart
type_list = []
for ctype in tfbct_df["type"]:
    type_list.append(ctype)
# print(type_list) # For testing only

pct_list = []
for fpct in tfbct_df["pct"]:
    pct_list.append(fpct)
# print(pct_list) # For testing only

# Build the Pie Chart - Create the data
# Labels for the sections of our pie chart
labels = type_list
# print(labels) # For testing only

# The values of each section of the pie chart
sizes = pct_list
# print(sizes) # For testing only

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]
# colors # For testing only
          
# Tells matplotlib to seperate the "Python" section from the others
explode = [0, 0, 0.1]

# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal") 

# Add a title
plt.title("% of Total Fares by City Type", pad=15, fontsize=18)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=180)
plt.show()

# Save Figure
plt.savefig("faresbycitytype.png")
```

## Total Rides by City Type

Three Observable Trends, Comments, & Questions for Future Data Analysis/Research:
--------------------------------------------------------------------------------------------------------------------------

1. In the Pyber Ride Sharing Data (2018) scatter plot, Urban cities are clustered towards more total rides and lower average fares per ride. They also have the largest number of drivers (as shown by circle size). Suburban cities are clustered towards the middle of the plot regarding total rides and fares. Rural cities are clustered towards a lower number of total rides (is this due to fewer drivers in Rural areas or lack of demand for rides) and higher average fares and have the fewest number of drvers. Only 2.6% of the company's drivers work in Rural areas whereas 80.9% of our drivers work in Urban areas. The avreage number of drivers in Rural areas is 4.3 per city whereas the average is 36.4 drivers in Urbzn cities.

From the Bubble Plot of Ride Sharing Data analysis section (reproduced here):
  There are 2,405 Urban drivers. Their numbers range from 3 to 73 per city with an average of 36.4.
  There are 490 Suburban drivers. Their numbers range from 1 to 25 per city with an average of 13.6.
  There are 78 Rural drivers. Their numbers range from 1 to 9 per city with an average of 4.3.

2. Are the lower average fares in Urban cities due to the higher number of drivers in those cities and thus the competition? Likewise, are the higher average fares in Rural cities due to the lack of supply of drivers in those cities? However, in Rural cities I would expect the lack of demand for rides to moderate (lessen the demand of) the higher fares.

3. Do drivers in Rural areas (with higher average fares) make higher annual incomes than drivers in Urban areas (with lower average fares)?  Or is it a wash? In other words, do the higher fares but fewer rides in Rural areas balance out with lower fares and many more rides in Urban areas? This would be a good question to be answered by some statistical analysis.
